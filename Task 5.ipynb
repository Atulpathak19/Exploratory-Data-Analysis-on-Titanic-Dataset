```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("train.csv")  # Make sure file is in same folder

# Show first 5 rows
df.head()
```

```python
df.info()
df.describe()
df.describe(include='object')
df.isnull().sum()
```

```python
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())
```

```python
sns.pairplot(df[['Survived','Pclass','Age','SibSp','Parch','Fare']])
plt.show()

plt.figure(figsize=(8,5))
sns.heatmap(df[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()

sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
print("""
Observations:
1. Females had higher survival rates.
2. 1st class passengers survived more than lower classes.
3. Higher fares linked to survival.
4. Children had better chances than adults.
""")
```
